```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

## ARMA(0,1)

```python
arparams = np.array([0 ])
maparams = np.array([np.random.uniform(0, .5)])
arma_process = sm.tsa.ArmaProcess.from_coeffs(arparams, maparams)
y = arma_process.generate_sample(100)

fig,(ax1,ax2,ax3) = plt.subplots(1,3, figsize = (20,5))
sns.lineplot(x = range(0,100), y = y, ax=ax1)
sm.graphics.tsa.plot_acf(y,lags =20, ax = ax2)
sm.graphics.tsa.plot_pacf(y, lags=20, method="ywm", ax = ax3);
```
